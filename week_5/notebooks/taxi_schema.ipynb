```python
years = [2020,2021]
types = ['yellow','green']
months = [x for x in range(1,13)]
sche = [yellow_schema,green_schema]

for typ,sch in zip(types,sche):
    for year in years:
        #try:
        for month in months:
            print(f"processing for {typ}/{year}/{month:02d}")



            input_path = f"{path}/{typ}/{year}/{month:02d}/"
            output_path = f"{path_2}/{typ}/{year}/{month:02d}/"

            df = spark.read.\
                     option("header", "true").\
                     schema(sch).\
                     csv(input_path)


            df.repartition(4).write.mode("ignore").parquet(output_path)

            print(f"conversion completed for {typ}/{year}/{month:02d}")
        #except ps.sql.utils.AnalysisException:
            #pass
```
